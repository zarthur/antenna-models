## Generating the Geometry

To model the antenna, we'll need to define a geometry. I'll use [4nec2](https://www.qsl.net/4nec2/) to help validate the generated geometry.

```python
from support import generate_antenna_geometry_4nec2, generate_nec_geometry_lines
```

```python
freq = 1_900_000  # 1.9 MHz, 160 Meters
geometry = generate_antenna_geometry_4nec2(freq, pipe_radius=0.25, segments_per_turn=9, wire_radius=0.01)
```

```python
geometry
```

```python
geometry_nec = generate_nec_geometry_lines(geometry)
print(''.join(geometry_nec[:5]))
```

```python
# need to fix these
prefix = [
    'CM Coiled 160 M Antenna\n',
    'CE\n'
]

suffix = [
    'GE 0\n',
    'GN 2 0 0 0 13 .005\n',
    'LD 5 0 0 0 58000000	\' Copper\n',
    'LD 0 3  5  5 0 0 36.84pF\n',
    'FR 0 1 0 0 14.05 1\n',
    'EX 0 1  5  0  1.0 0.0\n',
    'EN\n'
]

```

```python
filename = 'coil-160m.nec'

with open(filename, 'w') as outfile:
    for section in [prefix, geometry_nec, suffix]:
        outfile.writelines(section)
```
